# Merge all downscaled variable tables to one single table

```python
import os, sys
from pathlib import Path

import pandas as pd
from functools import partial
import glob

sys.path.append("../../") 
```

```python
RESULTS_PATH = Path("../../data/03-results/")
```

```python
data_tmax_df = pd.read_csv(RESULTS_PATH /  f"table_extracts/climate_atmosphere_downscaled_tmax.csv")
data_tmin_df = pd.read_csv(RESULTS_PATH /  f"table_extracts/climate_atmosphere_downscaled_tmin.csv")
data_tave_df = pd.read_csv(RESULTS_PATH /  f"table_extracts/climate_atmosphere_downscaled_tave.csv")
data_pr_df = pd.read_csv(RESULTS_PATH / f"table_extracts/climate_atmosphere_downscaled_precip.csv").rename(columns={'precip':'pr'})
```

```python
data_tmax_df['tmax'] = data_tmax_df['tmax'].round(2)
data_tmin_df['tmin'] = data_tmin_df['tmin'].round(2)
data_tave_df['tave'] = data_tave_df['tave'].round(2)
data_pr_df['pr'] = data_pr_df['pr'].round(2)
```

```python
data_brgy_df = data_tmin_df.merge(data_tmax_df, on=['date','adm4_pcode'], how='outer')
data_brgy_df = data_brgy_df.merge(data_tave_df, on=['date','adm4_pcode'], how='outer')
data_brgy_df = data_brgy_df.merge(data_pr_df, on=['date','adm4_pcode'], how='outer')
data_brgy_df
```

```python
data_brgy_df.to_csv(RESULTS_PATH / f"table_extracts/climate_atmosphere_downscaled.csv", index=False)
```
